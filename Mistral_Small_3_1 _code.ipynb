```python
import timeit
from memory_profiler import memory_usage
from typing import List

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        def f(i: int, j: int, k: int) -> int:
            if i >= m:
                return nums2[j + k - 1]
            if j >= n:
                return nums1[i + k - 1]
            if k == 1:
                return min(nums1[i], nums2[j])
            p = k // 2
            x = nums1[i + p - 1] if i + p - 1 < m else float('inf')
            y = nums2[j + p - 1] if j + p - 1 < n else float('inf')
            return f(i + p, j, k - p) if x < y else f(i, j + p, k - p)

        m, n = len(nums1), len(nums2)
        a = f(0, 0, (m + n + 1) // 2)
        b = f(0, 0, (m + n + 2) // 2) if (m + n) % 2 == 0 else a
        return (a + b) / 2

solution = Solution()

# Test cases with expected outputs
test_cases = [
    ([20, 67], [37, 85], 52.0),
    ([1, 2, 13, 22, 34, 46, 63, 86], [59, 80], 40.0),
    ([8, 57, 82, 87], [8, 18, 20, 23, 40, 41, 54, 63, 72, 93], 47.5),
    ([3, 36, 78], [13, 20, 28, 45, 59, 89], 36.0),
    ([17, 34, 43, 48, 51, 53, 83, 88], [43, 48, 54, 78, 88], 51.0),
    ([21, 28, 53, 63, 94], [6, 9, 25, 27, 31, 55, 60, 67, 78], 42.0),
    ([17, 50, 54], [1, 20, 46, 52, 53, 62, 63, 66, 70], 52.5),
    ([11, 13, 15, 27, 29, 55, 74, 89, 96], [9, 39, 72], 34.0),
    ([21, 24, 26, 33, 50, 53, 68, 87, 100], [1, 6, 12, 14, 34, 64, 81, 92, 93], 42.0),
    ([12, 29, 31, 38, 45, 47, 73, 76, 89, 92], [2, 22, 57, 85], 46.0)
]

def run_test_case(nums1, nums2, expected_result):
    # Measure the execution time
    start_time = timeit.default_timer()
    result = solution.findMedianSortedArrays(nums1, nums2)
    end_time = timeit.default_timer()
    execution_time = end_time - start_time

    # Measure the memory usage in bytes (memory_profiler reports in MB by default)
    mem_usage_func = lambda: solution.findMedianSortedArrays(nums1, nums2)
    mem_usage_series = memory_usage(mem_usage_func, interval=0.1)
    mem_usage = [mu * 1024 * 1024 for mu in mem_usage_series]  # Convert MB to bytes

    max_memory_usage = max(mem_usage)
    total_memory_usage = sum(mem_usage)
    average_memory_usage = total_memory_usage / len(mem_usage) if mem_usage else 0

    # Result comparison
    passed = result == expected_result

    # Collect results
    return (passed, result, execution_time, max_memory_usage, average_memory_usage, total_memory_usage)

# Collecting performance data
results = []
for nums1, nums2, expected_result in test_cases:
    passed, result, et, mu_max, mu_avg, tmu = run_test_case(nums1, nums2, expected_result)
    results.append((nums1, nums2, expected_result, passed, result, et, mu_max, mu_avg, tmu))

# Output each result and metric
for index, (nums1, nums2, expected_result, passed, result, et, mu_max, mu_avg, tmu) in enumerate(results):
    print(f"Test Case {index + 1}:")
    print(f"  Input: nums1 = {nums1}, nums2 = {nums2}")
    print(f"  Expected Result: {expected_result}")
    print(f"  Passed: {passed}")
    print(f"  Actual Result: {result} (if not passed)")
    print(f"  Execution Time (ET): {et:.6f} seconds")
    print(f"  Max Memory Usage (MU): {mu_max:.6f} bytes")
    print(f"  Total Memory Usage (TMU): {tmu:.6f} bytes")
    print(f"  Average Memory Usage: {mu_avg:.6f} bytes")
    print("-" * 50)

# Calculate overall metrics
execution_times = [res[5] for res in results]
max_memory_usages = [res[6] for res in results]
average_memory_usages = [res[7] for res in results]
total_memory_usages = [res[8] for res in results]

average_execution_time = sum(execution_times) / len(execution_times)
normalized_execution_time = 1  # Just an example, actual implementation may vary
overall_max_memory_usage = max(max_memory_usages)
normalized_max_memory_usage = overall_max_memory_usage  # Just an example, actual implementation may vary
total_memory_usage = sum(total_memory_usages)
normalized_total_memory_usage = total_memory_usage  # Just an example, actual implementation may vary
average_memory_usage = sum(average_memory_usages) / len(average_memory_usages)

# Output the aggregated results
print("\nSummary of Performance Metrics:")
print(f"Average Execution Time (ET): {average_execution_time:.6f} seconds")
print(f"Normalized Execution Time (NET): {normalized_execution_time:.6f}")
print(f"Overall Max Memory Usage (MU): {overall_max_memory_usage:.6f} bytes")
print(f"Normalized Max Memory Usage (NMU): {normalized_max_memory_usage:.6f} bytes")
print(f"Total Memory Usage (TMU): {total_memory_usage:.6f} bytes")
print(f"Normalized Total Memory Usage (NTMU): {normalized_total_memory_usage:.6f} bytes")
print(f"Average Memory Usage: {average_memory_usage:.6f} bytes")
```

```python
import time
import tracemalloc
from typing import List

class Solution:
    def minOperations(self, nums: List[int]) -> int:
        n = len(nums)
        nums = sorted(set(nums))
        ans, j = n, 0
        for i, v in enumerate(nums):
            while j < len(nums) and nums[j] - v <= n - 1:
                j += 1
            ans = min(ans, n - (j - i))
        return ans

def run_test_cases(test_cases: List[tuple]):
    results = []
    execution_times = []
    max_memory_uses = []
    total_memory_uses = []
    average_memory_usages = []

    for i, (nums, expected) in enumerate(test_cases):
        print(f"Running test case {i+1}/{len(test_cases)}")
        tracemalloc.start()

        start_time = time.time()
        solution = Solution()
        result = solution.minOperations(nums)
        end_time = time.time()

        snapshot = tracemalloc.take_snapshot()
        tracemalloc.stop()

        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.reset_peak()

        total_memory_usage = sum(stat.size for stat in snapshot.statistics('lineno'))
        max_memory_usage = snapshot.statistics('lineno')[0].size if snapshot.statistics('lineno') else 0
        average_memory_usage = total_memory_usage / len(snapshot.statistics('lineno')) if snapshot.statistics('lineno') else 0

        execution_times.append(end_time - start_time)
        max_memory_uses.append(max_memory_usage)
        total_memory_uses.append(total_memory_usage)
        average_memory_usages.append(average_memory_usage)

        results.append({
            "test_case": i + 1,
            "execution_time_ms": (end_time - start_time) * 1000,
            "max_memory_usage": max_memory_usage,
            "average_memory_usage": average_memory_usage,
            "total_memory_usage": total_memory_usage,
            "peak_memory_usage": peak,
            "output": result,
            "expected": expected
        })

        assert result == expected, f"Test case {i+1} failed: expected {expected}, got {result}"

    # Calculate overall metrics
    overall_metrics = calculate_overall_metrics(results, execution_times, max_memory_uses, total_memory_uses, average_memory_usages)
    normalized_results = normalize_results(results, overall_metrics)

    return normalized_results, overall_metrics

def normalize_results(results, overall_metrics):
    for result in results:
        result['normalized_execution_time'] = result['execution_time_ms'] / overall_metrics['max_execution_time']
        result['normalized_max_memory_usage'] = result['max_memory_usage'] / overall_metrics['max_max_memory_usage']
        result['normalized_total_memory_usage'] = result['total_memory_usage'] / overall_metrics['max_total_memory_usage']
    return results

def calculate_overall_metrics(results, execution_times, max_memory_uses, total_memory_uses, average_memory_usages):
    max_execution_time = max(execution_times)
    max_max_memory_usage = max(max_memory_uses)
    max_total_memory_usage = max(total_memory_uses)
    average_execution_time = sum(execution_times) / len(execution_times)
    average_max_memory_usage = sum(max_memory_uses) / len(max_memory_uses)
    average_total_memory_usage = sum(total_memory_uses) / len(total_memory_uses)
    overall_average_memory_usage = sum(average_memory_usages) / len(average_memory_usages)

    return {
        'max_execution_time': max_execution_time,
        'max_max_memory_usage': max_max_memory_usage,
        'max_total_memory_usage': max_total_memory_usage,
        'average_execution_time': average_execution_time,
        'average_max_memory_usage': average_max_memory_usage,
        'average_total_memory_usage': average_total_memory_usage,
        'overall_average_memory_usage': overall_average_memory_usage
    }

# Test cases to run
test_cases = [
    ([35, 62], 1),
    ([7, 1, 49], 2),
    ([18, 22, 71], 2),
    ([9, 90, 34, 55, 84, 57, 35, 98, 81], 7),
    ([88, 59, 66, 75, 84], 3),
    ([72, 65, 31, 86, 48, 15, 13, 61, 92, 47], 8),
    ([86, 52, 37, 33, 99, 67], 4),
    ([95, 84, 52, 70, 9], 4),
    ([14, 27, 88, 49, 80, 24, 2, 39, 71], 7),
    ([95, 74, 83, 46, 86, 7, 67], 5)
]

# Execute the test cases
normalized_results, overall_metrics = run_test_cases(test_cases)

# Print the normalized results
for result in normalized_results:
    print(f"Test case {result['test_case']}:")
    print(f"  Output: {result['output']}")
    print(f"  Expected: {result['expected']}")
    print(f"  Execution Time (ET): {result['execution_time_ms']:.4f} ms")
    print(f"  Normalized Execution Time (NET): {result['normalized_execution_time']:.4f}")
    print(f"  Max Memory Usage (MU): {result['max_memory_usage']} bytes")
    print(f"  Normalized Max Memory Usage (NMU): {result['normalized_max_memory_usage']:.4f}")
    print(f"  Total Memory Usage (TMU): {result['total_memory_usage']} bytes")
    print(f"  Normalized Total Memory Usage (NTMU): {result['normalized_total_memory_usage']:.4f}")
    print(f"  Average Memory Usage: {result['average_memory_usage']:.4f} bytes")
    print(f"  Peak Memory Usage: {result['peak_memory_usage']} bytes")
    print()

# Print overall metrics
print("Overall Metrics:")
print(f"  Average Execution Time (ET): {overall_metrics['average_execution_time']:.6f} seconds")
print(f"  Overall Maximum Memory Usage (Overall MU): {overall_metrics['max_max_memory_usage']} bytes")
print(f"  Average Maximum Memory Usage (Average MU): {overall_metrics['average_max_memory_usage']} bytes")
print(f"  Total Memory Usage (TMU): {overall_metrics['average_total_memory_usage']:.6f} bytes")
print(f"  Overall Average Memory Usage: {overall_metrics['overall_average_memory_usage']:.6f} bytes")
```
