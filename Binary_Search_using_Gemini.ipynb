```python
import time
from typing import List
from math import inf
from memory_profiler import memory_usage

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        def f(i: int, j: int, k: int) -> int:
            if i >= m:
                return nums2[j + k - 1]
            if j >= n:
                return nums1[i + k - 1]
            if k == 1:
                return min(nums1[i], nums2[j])
            p = k // 2
            x = nums1[i + p - 1] if i + p - 1 < m else inf
            y = nums2[j + p - 1] if j + p - 1 < n else inf
            return f(i + p, j, k - p) if x < y else f(i, j + p, k - p)

        m, n = len(nums1), len(nums2)
        a = f(0, 0, (m + n + 1) // 2)
        b = f(0, 0, (m + n + 2) // 2)
        return (a + b) / 2

if __name__ == "__main__":
    solution = Solution()
    test_cases = [
        ([20, 67], [37, 85], 52.0),
        ([1, 2, 13, 22, 34, 46, 63, 86], [59, 80], 40.0),
        ([8, 57, 82, 87], [8, 18, 20, 23, 40, 41, 54, 63, 72, 93], 47.5),
        ([3, 36, 78], [13, 20, 28, 45, 59, 89], 36.0),
        ([17, 34, 43, 48, 51, 53, 83, 88], [43, 48, 54, 78, 88], 51.0),
        ([21, 28, 53, 63, 94], [6, 9, 25, 27, 31, 55, 60, 67, 78], 42.0),
        ([17, 50, 54], [1, 20, 46, 52, 53, 62, 63, 66, 70], 52.5),
        ([11, 13, 15, 27, 29, 55, 74, 89, 96], [9, 39, 72], 34.0),
        ([21, 24, 26, 33, 50, 53, 68, 87, 100], [1, 6, 12, 14, 34, 64, 81, 92, 93], 42.0),
        ([12, 29, 31, 38, 45, 47, 73, 76, 89, 92], [2, 22, 57, 85], 46.0),
    ]

    execution_times = []
    memory_usages_list = []

    for nums1, nums2, expected in test_cases:
        start_time = time.time()
        mem_usage = memory_usage((solution.findMedianSortedArrays, (nums1, nums2)), retval=True)
        result = mem_usage[1]  # The return value is the second element
        max_mem = max(mem_usage[0]) # Get the maximum memory usage from the list
        end_time = time.time()
        execution_times.append(end_time - start_time)
        memory_usages_list.append(max_mem)

        print(f"nums1={nums1}, nums2={nums2}, Output={result}, Expected={expected}, Pass={abs(result - expected) < 1e-9}, Max Memory Usage={max_mem * 1024 * 1024:.2f} bytes")

    avg_execution_time = sum(execution_times) / len(execution_times)
    max_execution_time = max(execution_times)
    min_execution_time = min(execution_times)

    normalized_execution_times = [(et - min_execution_time) / (max_execution_time - min_execution_time) if max_execution_time > min_execution_time else 0 for et in execution_times]
    avg_normalized_execution_time = sum(normalized_execution_times) / len(normalized_execution_times)

    overall_max_memory_usage_bytes = max(memory_usages_list) * 1024 * 1024
    overall_avg_memory_usage_bytes = (sum(memory_usages_list) / len(memory_usages_list)) * 1024 * 1024
    total_memory_usage_bytes = sum(memory_usages_list) * 1024 * 1024
    min_memory_usage_mb = min(memory_usages_list)
    max_memory_usage_mb = max(memory_usages_list)

    normalized_max_memory_usages = [(mu - min_memory_usage_mb) / (max_memory_usage_mb - min_memory_usage_mb) if max_memory_usage_mb > min_memory_usage_mb else 0 for mu in memory_usages_list]
    avg_normalized_memory_usage = sum(normalized_max_memory_usages) / len(normalized_max_memory_usages)
    normalized_total_memory_usage = (total_memory_usage_bytes - min_memory_usage_mb * len(memory_usages_list) * 1024 * 1024) / (max_memory_usage_mb * len(memory_usages_list) * 1024 * 1024 - min_memory_usage_mb * len(memory_usages_list) * 1024 * 1024) if max_memory_usage_mb > min_memory_usage_mb else 0


    print("\nMetrics:")
    print(f"Execution Times (ET): {execution_times}")
    print(f"Average Execution Time: {avg_execution_time:.6f} seconds")
    print(f"Normalized Execution Times (NET): {normalized_execution_times}")
    print(f"Average Normalized Execution Time (Average NET): {avg_normalized_execution_time:.6f}")
    print(f"Overall Maximum Memory Usage: {overall_max_memory_usage_bytes:.2f} bytes")
    print(f"Overall Average Memory Usage: {overall_avg_memory_usage_bytes:.2f} bytes")
    print(f"Normalized Max Memory Usage (NMU): {normalized_max_memory_usages[-1]:.6f}")
    print(f"Total Memory Usage (TMU): {total_memory_usage_bytes:.2f} bytes")
    print(f"Normalized Total Memory Usage (NTMU): {normalized_total_memory_usage:.6f}")
    print(f"Average Memory Usage (per test case): {(sum(memory_usages_list) / len(memory_usages_list)) * 1024 * 1024:.2f} bytes")
```

```python
import time
from typing import List
from memory_profiler import memory_usage
import random

class Solution:
    def minimumBoxes(self, n: int) -> int:
        if n <= 0:
            return 0

        low = 1
        high = 2 * (n**0.5)
        k = 0

        while low <= high:
            mid = (low + high) // 2
            total_boxes = mid * (mid + 1) * (mid + 2) // 6
            if total_boxes >= n:
                k = mid
                high = mid - 1
            else:
                low = mid + 1

        floor_count = (k - 1) * k // 2
        remaining_boxes = n - (k - 1) * k * (k + 1) // 6

        l = 1
        while remaining_boxes > 0:
            floor_count += 1
            remaining_boxes -= l
            l += 1

        return floor_count

def run_test_case(n: int, expected: int):
    solution = Solution()
    start_time = time.time()
    mem_usage = memory_usage((solution.minimumBoxes, (n,)), retval=True)
    end_time = time.time()
    execution_time = end_time - start_time
    max_memory_usage_bytes = max(mem_usage[0]) * 1024 if mem_usage[0] else 0.0
    total_memory_usage_bytes = sum(mem_usage[0]) * 1024 if mem_usage[0] else 0.0
    average_memory_usage_bytes = total_memory_usage_bytes / len(mem_usage[0]) if mem_usage[0] else 0.0
    result = mem_usage[1]
    passed = result == expected
    return execution_time, max_memory_usage_bytes, total_memory_usage_bytes, average_memory_usage_bytes, result, passed

if __name__ == "__main__":
    test_cases = [
        (459320720, 982324),
        (431102999, 941659),
        (573810324, 1139286),
        (703979226, 1306040),
        (569073436, 1133124),
        (993326376, 1643056),
        (238461863, 634399),
        (135673094, 435568),
        (850598996, 1481505),
        (918706813, 1559722),
    ]

    execution_times = []
    max_memory_usages_bytes = []
    total_memory_usages_bytes = []
    average_memory_usages_bytes = []
    pass_statuses = []
    actual_outputs = []
    expected_outputs = [tc[1] for tc in test_cases]

    for i, (n, expected) in enumerate(test_cases):
        et, mu_bytes, tmu_bytes, avg_mu_bytes, output, passed = run_test_case(n, expected)
        execution_times.append(et)
        max_memory_usages_bytes.append(mu_bytes)
        total_memory_usages_bytes.append(tmu_bytes)
        average_memory_usages_bytes.append(avg_mu_bytes)
        actual_outputs.append(output)
        pass_statuses.append(passed)
        print(f"n={n}, Output={output}, Expected={expected}, Pass={passed}, ET={et:.6f} s, MU={mu_bytes:.2f} bytes, Avg MU={avg_mu_bytes:.2f} bytes")

    avg_et = sum(execution_times) / len(execution_times) if execution_times else 0
    min_et = min(execution_times) if execution_times else float('inf')
    max_et = max(execution_times) if execution_times else float('-inf')
    normalized_et = [(et - min_et) / (max_et - min_et) if (max_et - min_et) > 0 else 0 for et in execution_times]

    avg_mu_bytes = sum(max_memory_usages_bytes) / len(max_memory_usages_bytes) if max_memory_usages_bytes else 0
    min_mu_bytes = min(max_memory_usages_bytes) if max_memory_usages_bytes else float('inf')
    max_mu_bytes = max(max_memory_usages_bytes) if max_memory_usages_bytes else float('-inf')
    normalized_mu = [(mu - min_mu_bytes) / (max_mu_bytes - min_mu_bytes) if (max_mu_bytes - min_mu_bytes) > 0 else 0 for mu in max_memory_usages_bytes]

    avg_avg_mu_bytes = sum(average_memory_usages_bytes) / len(average_memory_usages_bytes) if average_memory_usages_bytes else 0
    min_avg_mu_bytes = min(average_memory_usages_bytes) if average_memory_usages_bytes else float('inf')
    max_avg_mu_bytes = max(average_memory_usages_bytes) if average_memory_usages_bytes else float('-inf')
    normalized_avg_mu = [(avg_mu - min_avg_mu_bytes) / (max_avg_mu_bytes - min_avg_mu_bytes) if (max_avg_mu_bytes - min_avg_mu_bytes) > 0 else 0 for avg_mu in average_memory_usages_bytes]

    accuracy = sum(pass_statuses) / len(pass_statuses) if pass_statuses else 0

    print("\nMetrics:")
    print(f"Average Execution Time: {avg_et:.6f} seconds")
    print(f"Max Execution Time: {max_et:.6f} seconds")
    print(f"Min Execution Time: {min_et:.6f} seconds")
    print(f"Normalized Execution Times: {normalized_et}")
    print(f"Average Max Memory Usage: {avg_mu_bytes:.2f} bytes")
    print(f"Max Memory Usage: {max_mu_bytes:.2f} bytes")
    print(f"Min Memory Usage: {min_mu_bytes:.2f} bytes")
    print(f"Normalized Max Memory Usages: {normalized_mu}")
    print(f"Average Average Memory Usage: {avg_avg_mu_bytes:.2f} bytes")
    print(f"Max Average Memory Usage: {max_avg_mu_bytes:.2f} bytes")
    print(f"Min Average Memory Usage: {min_avg_mu_bytes:.2f} bytes")
    print(f"Normalized Average Memory Usages: {normalized_avg_mu}")
    print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import time
from typing import List
from memory_profiler import memory_usage

class Solution:
    def minOperations(self, nums: List[int]) -> int:
        n = len(nums)
        unique_sorted_nums = sorted(list(set(nums)))
        min_ops = n

        for i in range(len(unique_sorted_nums)):
            start = unique_sorted_nums[i]
            end = start + n - 1
            count_within_range = 0
            for num in unique_sorted_nums:
                if start <= num <= end:
                    count_within_range += 1
            operations = n - count_within_range
            min_ops = min(min_ops, operations)

        return min_ops

def run_test_case(nums: List[int], expected: int):
    solution = Solution()
    start_time = time.time()
    mem_usage = memory_usage((solution.minOperations, (nums,)), retval=True)
    end_time = time.time()
    execution_time = end_time - start_time
    max_memory_usage_bytes = max(mem_usage[0]) * 1024 if mem_usage[0] else 0.0  # Convert MB to bytes
    total_memory_usage_bytes = sum(mem_usage[0]) * 1024 if mem_usage[0] else 0.0  # Convert MB to bytes
    average_memory_usage_bytes = total_memory_usage_bytes / len(mem_usage[0]) if mem_usage[0] else 0.0
    result = mem_usage[1]
    passed = result == expected
    return execution_time, max_memory_usage_bytes, total_memory_usage_bytes, average_memory_usage_bytes, result, passed

if __name__ == "__main__":
    test_cases = [
        ([35, 62], 1),
        ([7, 1, 49], 2),
        ([18, 22, 71], 2),
        ([9, 90, 34, 55, 84, 57, 35, 98, 81], 7),
        ([88, 59, 66, 75, 84], 3),
        ([72, 65, 31, 86, 48, 15, 13, 61, 92, 47], 8),
        ([86, 52, 37, 33, 99, 67], 4),
        ([95, 84, 52, 70, 9], 4),
        ([14, 27, 88, 49, 80, 24, 2, 39, 71], 7),
        ([95, 74, 83, 46, 86, 7, 67], 5),
    ]

    execution_times = []
    max_memory_usages_bytes = []
    total_memory_usages_bytes = []
    average_memory_usages_bytes = []
    pass_statuses = []
    actual_outputs = []
    expected_outputs = [tc[1] for tc in test_cases]

    for i, (nums, expected) in enumerate(test_cases):
        et, mu_bytes, tmu_bytes, avg_mu_bytes, output, passed = run_test_case(nums, expected)
        execution_times.append(et)
        max_memory_usages_bytes.append(mu_bytes)
        total_memory_usages_bytes.append(tmu_bytes)
        average_memory_usages_bytes.append(avg_mu_bytes)
        actual_outputs.append(output)
        pass_statuses.append(passed)
        print(f"nums={nums}, Output={output}, Expected={expected}, Pass={passed}, ET={et:.6f} s, MU={mu_bytes:.2f} bytes, TMU={tmu_bytes:.2f} bytes, Avg MU={avg_mu_bytes:.2f} bytes")

    avg_et = sum(execution_times) / len(execution_times) if execution_times else 0
    min_et = min(execution_times) if execution_times else float('inf')
    max_et = max(execution_times) if execution_times else float('-inf')
    normalized_et = [(et - min_et) / (max_et - min_et) if (max_et - min_et) > 0 else 0 for et in execution_times]

    avg_mu_bytes = sum(max_memory_usages_bytes) / len(max_memory_usages_bytes) if max_memory_usages_bytes else 0
    min_mu_bytes = min(max_memory_usages_bytes) if max_memory_usages_bytes else float('inf')
    max_mu_bytes = max(max_memory_usages_bytes) if max_memory_usages_bytes else float('-inf')
    normalized_mu = [(mu - min_mu_bytes) / (max_mu_bytes - min_mu_bytes) if (max_mu_bytes - min_mu_bytes) > 0 else 0 for mu in max_memory_usages_bytes]

    avg_tmu_bytes = sum(total_memory_usages_bytes) / len(total_memory_usages_bytes) if total_memory_usages_bytes else 0
    min_tmu_bytes = min(total_memory_usages_bytes) if total_memory_usages_bytes else float('inf')
    max_tmu_bytes = max(total_memory_usages_bytes) if total_memory_usages_bytes else float('-inf')
    normalized_tmu = [(tmu - min_tmu_bytes) / (max_tmu_bytes - min_tmu_bytes) if (max_tmu_bytes - min_tmu_bytes) > 0 else 0 for tmu in total_memory_usages_bytes]

    avg_avg_mu_bytes = sum(average_memory_usages_bytes) / len(average_memory_usages_bytes) if average_memory_usages_bytes else 0
    min_avg_mu_bytes = min(average_memory_usages_bytes) if average_memory_usages_bytes else float('inf')
    max_avg_mu_bytes = max(average_memory_usages_bytes) if average_memory_usages_bytes else float('-inf')
    normalized_avg_mu = [(avg_mu - min_avg_mu_bytes) / (max_avg_mu_bytes - min_avg_mu_bytes) if (max_avg_mu_bytes - min_avg_mu_bytes) > 0 else 0 for avg_mu in average_memory_usages_bytes]

    accuracy = sum(pass_statuses) / len(pass_statuses) if pass_statuses else 0

    print("\nMetrics:")
    print(f"Average Execution Time: {avg_et:.6f} seconds")
    print(f"Max Execution Time: {max_et:.6f} seconds")
    print(f"Min Execution Time: {min_et:.6f} seconds")
    print(f"Normalized Execution Times: {normalized_et}")
    print(f"Average Max Memory Usage: {avg_mu_bytes:.2f} bytes")
    print(f"Max Memory Usage: {max_mu_bytes:.2f} bytes")
    print(f"Min Memory Usage: {min_mu_bytes:.2f} bytes")
    print(f"Normalized Max Memory Usages: {normalized_mu}")
    print(f"Average Total Memory Usage: {avg_tmu_bytes:.2f} bytes")
    print(f"Max Total Memory Usage: {max_tmu_bytes:.2f} bytes")
    print(f"Min Total Memory Usage: {min_tmu_bytes:.2f} bytes")
    print(f"Normalized Total Memory Usages: {normalized_tmu}")
    print(f"Average Average Memory Usage: {avg_avg_mu_bytes:.2f} bytes")
    print(f"Max Average Memory Usage: {max_avg_mu_bytes:.2f} bytes")
    print(f"Min Average Memory Usage: {min_avg_mu_bytes:.2f} bytes")
    print(f"Normalized Average Memory Usages: {normalized_avg_mu}")
    print(f"Accuracy: {accuracy * 100:.2f}%")
```
